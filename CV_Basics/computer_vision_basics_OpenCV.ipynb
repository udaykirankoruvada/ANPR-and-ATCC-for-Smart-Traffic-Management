```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Readig the image in Gray Scale 
# value are from 0 to 255

def main():
    _image = cv2.imread('image.jpg',0)

    # cv2.imshow("traffic image gray scale", _image)
    
    # # representation of image 
    image_matrix = np.array(_image)
    # print(image_matrix)

    # Matplotlib represetion 
    plt.imshow(image_matrix, cmap='gray')
    plt.colorbar(label='Pixel Intensity')
    plt.title("Grayscale Image Matrix Visualization")
    plt.axis('off')  # Optional: turn off axis labels for a cleaner view
    plt.show()

    cv2.waitKey(0)
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```

```python
# Readig the image in RGB

def main():
    _img = cv2.imread('image.jpg', cv2.IMREAD_COLOR)

    cv2.imshow("traffic image", _img)

    cv2.waitKey(0)
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```

```python
# Writig on images

import cv2
import numpy as np


def main():
    _img = cv2.imread('image.jpg', cv2.IMREAD_COLOR)
    
    # Coordinates for the text position
    # text_position = (20, _img.shape[0] - 400)
    text_position = (20, 50)

    
    cv2.putText(_img, "Car", text_position, 
                cv2.FONT_HERSHEY_SIMPLEX, 2.0, (255, 255, 255), 2)
    
    cv2.imshow("traffic image", _img)

    cv2.waitKey(0)
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```
